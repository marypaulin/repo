```python
# How many features can we now reasonably scale to?

df = []

compas = compas1.copy()

x_compas = compas.values[:,:-1]
y_compas = compas.values[:,-1]

X = np.copy(x_compas)

ridx = np.random.randint(6907, size=4000)
X[ridx,:] = np.random.randint(2, size=(len(ridx), X.shape[1]))

for j in range(12):
    
    _, _, _, _, nrule, ndata, totaltime, time_c, COUNT, C_c, accu = bbound(
        x_compas, y_compas, lamb=0.005, prior_metric="curiosity", MAXDEPTH=4)
    #compas = compas.append(compas)
    
    x_compas = np.hstack((x_compas,X[:,j].reshape(-1,1)))
    
    print(nrule, ndata, totaltime, time_c, COUNT, C_c, accu)
    df.append([nrule, ndata, totaltime, time_c, COUNT, C_c, accu])
        
df1 = pd.DataFrame(data=np.array(df), columns=["nrule", "ndata", "totaltime", "time_c", "COUNT", "C_c", "accu"])
df1.to_csv('../data/performance_compasbinary_feature.csv',index=False,sep=',')
```

```python
df = []

for i in range(len(idxs)):
    idx = idxs[i]
    compas = compas1.copy()
    for j in range(2,11):
        x_compas = compas.values[:,:-1]
        y_compas = compas.values[:,-1]
        _, _, _, _, nrule, ndata, totaltime, time_c, COUNT, C_c, accu = bbound(
            x_compas[:,idx], y_compas, lamb=0.005, prior_metric="curiosity", MAXDEPTH=4)
        compas = compas.append(compas1)
        
        df.append([nrule, ndata, totaltime, time_c, COUNT, C_c, accu])
        
df1 = pd.DataFrame(data=np.array(df), columns=["nrule", "ndata", "totaltime", "time_c", "COUNT", "C_c", "accu"])
df1.to_csv('../data/performance_compasbinary0.csv',index=False,sep=',')
```
