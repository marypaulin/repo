```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_la_T = pd.read_csv('5_6907_gini_0.001_4_True_.txt', delimiter=';')
df_la_F = pd.read_csv('5_6907_gini_0.001_4_False_.txt', delimiter=';')
```

```python
df_la_T.iloc[0:20]
```

```python
num = min(df_la_T.shape[0],df_la_F.shape[0])
num
```

```python
plt.figure(figsize=(16, 9))
plt.plot(df_la_T['#push'].iloc[0:60000], df_la_T['the_new_tree_depth'].iloc[0:60000])
plt.plot(df_la_F['#push'].iloc[0:60000], df_la_F['the_new_tree_depth'].iloc[0:60000])
plt.title('With(blue) and without(orange) lookahead bound, the depth of trees generated')
plt.xlabel('number of trees pushed into the queue')

plt.savefig("depth.jpg")
```

```python
plt.figure(figsize=(16, 9))
plt.plot(df_la_T['#push'].iloc[0:90000], df_la_T['R_c'].iloc[0:90000])
plt.plot(df_la_F['#push'].iloc[0:90000], df_la_F['R_c'].iloc[0:90000])
plt.title('With(blue) and without(orange) lookahead bound, best objective values')
plt.xlabel('number of trees pushed into the queue')

plt.savefig("best_objectives.jpg")
```

```python
plt.figure(figsize=(16, 9))
plt.plot(df_la_T['#push'], df_la_T['the_new_tree_lbound'])
plt.plot(df_la_F['#push'], df_la_F['the_new_tree_lbound'])
plt.title('With(blue) and without(orange) lookahead bound, the lower bound of trees generated')
plt.xlabel('number of trees pushed into the queue')

plt.savefig("lbound.jpg")
```

```python
plt.figure(figsize=(16, 9))
plt.plot(df_la_T['#push'], df_la_T['the_new_tree_objective'])
plt.plot(df_la_F['#push'], df_la_F['the_new_tree_objective'])
plt.title('With(blue) and without(orange) lookahead bound, the objective of trees generated')
plt.xlabel('number of trees pushed into the queue')

plt.savefig("objective.jpg")
```

```python
plt.figure(figsize=(16, 9))
plt.plot(range(df_la_T.shape[0]), df_la_T['queue_size'])
plt.plot(range(df_la_F.shape[0]), df_la_F['queue_size'])
plt.title('With(blue) and without(orange) lookahead bound, queue_sizes')
plt.xlabel('number of trees pushed into the queue')

plt.savefig("queue_sizes.jpg")
```
