```python
def plot_trace(data, title):

    plt.figure(num=None, figsize=(8, 6), facecolor='w', edgecolor='k')

    ax = plt.axes()
    
    ax.plot(data['#pop'], data['the_old_tree_lbound'],
            markersize=12, color = 'black', linestyle='--')
    ax.plot(data['#pop'], data['R_c'],
            markersize=12, color = 'blue')

    ax.plot(data['#pop'][np.argmin(data['R_c'])],min(data['R_c']),  '*', color = 'orange', markersize=18, alpha=0.75)

    ax.set_xlabel('Number of Trees Evaluated', fontsize=18)
    ax.set_ylabel('Value', fontsize=18)
    ax.set_title(title,fontsize=20,fontweight="bold")

    #plt.savefig("./trace_" + str(title) + ".pdf", bbox_inches='tight')

    plt.show()
```
