```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('../data/performance_compasbinary1.csv')
df2 = pd.read_csv('performance_USCensus1990_0025.csv')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X=np.array(df1.n).reshape([9,9]), 
                Y=np.array(df1.p).reshape([9,9]), 
                Z=np.array(df1.totaltime).reshape([9,9]),cmap=cm.coolwarm,)

ax.set_xlabel('n')
ax.set_ylabel('p')
ax.set_zlabel('Totoal Time')

plt.show()
fig.savefig('fig1a.png',dpi=300)
```

```python
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X=np.array(df1.n).reshape([9,9]), 
                Y=np.array(df1.p).reshape([9,9]), 
                Z=np.array(df1.timetooptimal).reshape([9,9]),cmap=cm.coolwarm,)

ax.set_xlabel('n')
ax.set_ylabel('p')
ax.set_zlabel('Time to Find the Optimal Tree')

plt.show()
fig.savefig('fig1b.png',dpi=300)
```

```python
# red dashes, blue squares and green triangles
fig = plt.figure(figsize=(9,6))
plt.plot(df1.p[df1.n==6907], df1.totaltime[df1.n==6907], '--', df1.p[df1.n==6907], df1.timetooptimal[df1.n==6907], 'r--')

plt.xlabel('Number of features p')
plt.ylabel('Time')
plt.title('Totoal Time and Time to Find the Optimal Tree for different p')
plt.show()
fig.savefig('fig1c.png',dpi=300)
```
