```python
def draw_accuracy(osdt_data, cart_data, binoct_data, fd, dataname):
    plt.figure(num=None, figsize=(8, 6), facecolor='w', edgecolor='k')

    ax = plt.axes()

    ax.plot(osdt_data['nleaves'][osdt_data['fold'] == fd], osdt_data['trainaccu_CART'][osdt_data['fold'] == fd], '>',
            markersize=12, color = 'blue', label='OSDT-Train')
    ax.plot(osdt_data['nleaves'][osdt_data['fold'] == fd], osdt_data['testaccu_CART'][osdt_data['fold'] == fd], '>',
            markersize=12, color = 'orange', label='OSDT-Test')

    ax.plot(cart_data['nleaves'][cart_data['fold'] == fd], cart_data['trainaccu_CART'][cart_data['fold'] == fd], '+',
            markersize=12, color = 'blue', label='CART-Train')
    ax.plot(cart_data['nleaves'][cart_data['fold'] == fd], cart_data['testaccu_CART'][cart_data['fold'] == fd], '+',
            markersize=12, color = 'orange', label='CART-Test')

    ax.plot(binoct_data['nleaves'][binoct_data['fold'] == fd], binoct_data['trainaccu'][binoct_data['fold'] == fd], 'x',
            markersize=12, color='blue', label='BinOCT-Train')
    ax.plot(binoct_data['nleaves'][binoct_data['fold'] == fd], binoct_data['testaccu'][binoct_data['fold'] == fd], 'x',
            markersize=12, color='orange', label='BinOCT-Test')

    ax.set_xlabel('Number of Leaves')
    ax.set_ylabel('Accuracy')
    ax.set_title('Classification Accuracy of ' + str(dataname) + ' Dataset ' + str(fd))

    #blue_line = mlines.Line2D([], [], color='black', marker='x',
    #                          markersize=12, label='Blue stars')
    #blue_patch = mpatches.Patch(color='blue', label='The training accuracy')
    #orange_patch = mpatches.Patch(color='orange', label='The testing accuracy')
    #ax.legend(handles=[blue_patch, orange_patch, blue_line], loc=4)

    ax.legend(loc='best')

    plt.savefig("./accuracy/" + str(dataname) + str(fd) + ".pdf", bbox_inches='tight')

    plt.show()
```

```python
# ONLY ONE FOLD
def draw_accuracy_train(osdt_data, cart_data, binoct_data, dataname):
    plt.figure(num=None, figsize=(8, 6), facecolor='w', edgecolor='k')

    ax = plt.axes()

    ax.plot(osdt_data['nleaves'][osdt_data['nleaves'] > 1], osdt_data['trainaccu_CART'][osdt_data['nleaves'] > 1], '>',
            markersize=12, color = 'blue', label='OSDT-Train')
    #ax.plot(osdt_data['nleaves'][osdt_data['fold'] == fd], osdt_data['testaccu_CART'][osdt_data['fold'] == fd], '>',
            #markersize=12, color = 'orange', label='OSDT-Test')

    ax.plot(cart_data['nleaves'][cart_data['nleaves'] > 1], cart_data['trainaccu_CART'][cart_data['nleaves'] > 1], '+',
            markersize=12, color = 'blue', label='CART-Train')
    #ax.plot(cart_data['nleaves'][cart_data['fold'] == fd], cart_data['testaccu_CART'][cart_data['fold'] == fd], '+',
            #markersize=12, color = 'orange', label='CART-Test')

    ax.plot(binoct_data['nleaves'][binoct_data['nleaves'] > 1], binoct_data['trainaccu'][binoct_data['nleaves'] > 1], 'x',
            markersize=12, color='blue', label='BinOCT-Train')
    #ax.plot(binoct_data['nleaves'][binoct_data['fold'] == fd], binoct_data['testaccu'][binoct_data['fold'] == fd], 'x',
            #markersize=12, color='orange', label='BinOCT-Test')
        
    ax.set_xlim(0, max(osdt_data['nleaves'])+2)
    

    ax.set_xlabel('Number of Leaves')
    ax.set_ylabel('Accuracy')
    ax.set_title('Classification Accuracy of ' + str(dataname) + ' Dataset ')

    #blue_line = mlines.Line2D([], [], color='black', marker='x',
    #                          markersize=12, label='Blue stars')
    #blue_patch = mpatches.Patch(color='blue', label='The training accuracy')
    #orange_patch = mpatches.Patch(color='orange', label='The testing accuracy')
    #ax.legend(handles=[blue_patch, orange_patch, blue_line], loc=4)

    ax.legend(loc='best')

    plt.savefig("./accuracy/" + str(dataname) + ".pdf", bbox_inches='tight')

    plt.show()
```

```python
# ONLY ONE FOLD, Cold and Warm
def draw_accuracy_cold_and_warm(osdt_data_cold, cart_data_cold, binoct_data_cold,
                                osdt_data_warm, cart_data_warm, binoct_data_warm, dataname, xlim, ylim0, ylim1):
    plt.figure(num=None, figsize=(9, 6), facecolor='w', edgecolor='k')

    ax = plt.axes()

    alp = 0.7

    #ax.plot(osdt_data_cold['nleaves'][osdt_data_cold['nleaves'] > 1],
    #        osdt_data_cold['trainaccu_CART'][osdt_data_cold['nleaves'] > 1], '>', alpha=alp,
    #        markersize=12, color='blue', label='OSDT-ColdStart')
    ax.plot(osdt_data_warm['nleaves'][osdt_data_warm['nleaves'] > 1],
            osdt_data_warm['trainaccu_CART'][osdt_data_warm['nleaves'] > 1], 'D', #alpha=alp,
            markersize=12, color='goldenrod', label='OSDT-WarmStart')

    ax.plot(cart_data_cold['nleaves'][cart_data_cold['nleaves'] > 1],
            cart_data_cold['trainaccu_CART'][cart_data_cold['nleaves'] > 1], 'D', #alpha=alp,
            markersize=12, color='mediumpurple', label='CART')

    #ax.plot(binoct_data_cold['nleaves'][binoct_data_cold['nleaves'] > 1],
    #        binoct_data_cold['trainaccu'][binoct_data_cold['nleaves'] > 1], 'x', alpha=alp,
    #        markersize=12, color='blue', label='BinOCT-ColdStart')
    ax.plot(binoct_data_warm['nleaves'][binoct_data_warm['nleaves'] > 1],
            binoct_data_warm['trainaccu'][binoct_data_warm['nleaves'] > 1], 'D', #alpha=alp,
            markersize=12, color='blue', label='BinOCT-WarmStart')

    ax.set_xlim(0, xlim)
    ax.set_ylim(ylim0, ylim1)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(18)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(18)
        
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    
    ax.axhline(y=max(osdt_data_warm['trainaccu_CART'][osdt_data_warm['nleaves'] < xlim]), color='grey', linestyle='--')

    ax.set_xlabel('Number of Leaves', fontsize=18)
    ax.set_ylabel('Accuracy', fontsize=18)
    ax.set_title('Classification Accuracy of ' + str(dataname) + ' Dataset ', fontsize=20,fontweight="bold")
    

    # blue_line = mlines.Line2D([], [], color='black', marker='x',
    #                          markersize=12, label='Blue stars')
    # blue_patch = mpatches.Patch(color='blue', label='The training accuracy')
    # orange_patch = mpatches.Patch(color='orange', label='The testing accuracy')
    # ax.legend(handles=[blue_patch, orange_patch, blue_line], loc=4)

    ax.legend(loc='best', fontsize=16)

    plt.savefig("./accuracy/" + str(dataname) + ".pdf", bbox_inches='tight')

    plt.show()
    
```
