```python
import matplotlib.pyplot as plt
```

```python
# Execution traces of queue sizes
fig = plt.figure(figsize=(16, 9))
sub1 = fig.add_subplot(221)
sub1.set_title('Priority Queue: Curiosity', fontsize=16)
sub1.set_xlabel('Time(s)')
sub1.set_ylabel('The Size of Queue')
sub1.plot(df1['time'], df1['queue_size'])
sub2 = fig.add_subplot(222)
sub2.set_title('Priority Queue: Bound', fontsize=16)
sub2.set_xlabel('Time(s)')
sub2.set_ylabel('The Size of Queue')
sub2.plot(df2['time'], df2['queue_size'])
sub3 = fig.add_subplot(223)
sub3.set_title('Priority Queue: Entropy', fontsize=16)
sub3.set_xlabel('Time(s)')
sub3.set_ylabel('The Size of Queue')
sub3.plot(df3['time'], df3['queue_size'])
sub4 = fig.add_subplot(224)
sub4.set_title('Priority Queue: Gini Index', fontsize=16)
sub4.set_xlabel('Time(s)')
sub4.set_ylabel('The Size of Queue')
sub4.plot(df4['time'], df4['queue_size'])
plt.tight_layout()
plt.show()
```

```python
# Execution traces of the best objective value and the accuracy
fig = plt.figure(figsize=(16, 9))
sub1 = fig.add_subplot(221)
sub1.set_title('Priority Queue: Curiosity', fontsize=16)
sub1.set_xlabel('Time(s)')
sub1.set_ylabel('The Best Objective and the Accuracy')
sub1.plot(df1['time'], df1['objective'])
sub1.plot(df1['time'], df1['accuracy'])
sub2 = fig.add_subplot(222)
sub2.set_title('Priority Queue: Bound', fontsize=16)
sub2.set_xlabel('Time(s)')
sub2.set_ylabel('The Best Objective and the Accuracy')
sub2.plot(df2['time'], df2['objective'])
sub2.plot(df2['time'], df2['accuracy'])
sub3 = fig.add_subplot(223)
sub3.set_title('Priority Queue: Entropy', fontsize=16)
sub3.set_xlabel('Time(s)')
sub3.set_ylabel('The Best Objective and the Accuracy')
sub3.plot(df3['time'], df3['objective'])
sub3.plot(df3['time'], df3['accuracy'])
sub4 = fig.add_subplot(224)
sub4.set_title('Priority Queue: Gini Index', fontsize=16)
sub4.set_xlabel('Time(s)')
sub4.set_ylabel('The Best Objective and the Accuracy')
sub4.plot(df4['time'], df4['objective'])
sub4.plot(df4['time'], df4['accuracy'])
plt.tight_layout()
plt.show()
```
